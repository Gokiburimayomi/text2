```python
import requests
import json
import time
import jieba 
token=input("請輸入權杖") 
res=requests.get("https://graph.facebook.com/v17.0/me/posts?access_token="+token) #讀取網址
jd=json.loads(res.text) #翻譯網頁#翻譯完的網頁存在變數 jd 
while jd["data"]!=[]: #當jd["data"]不等於[]
    time.sleep(1)  #等待一秒
    for post in jd["data"]: #把jd["data"]資料放進post
        if "message" in post:  #如果"message"在post裡面
            print(post["message"]) #放出message的內容
    res=requests.get(jd["paging"]['next']) # res=下一頁的網址
    jd=json.loads(res.text) 
```

```python
res=requests.get("https://graph.facebook.com/v17.0/me/posts?access_token="+token) #讀取網址
jd=json.loads(res.text) #翻譯網頁#翻譯完的網頁存在變數 jd
space = []
while jd["data"]!=[]: #當jd["data"]不等於[]
    time.sleep(2)  #等待二秒
    for post in jd["data"]: #把jd["data"]資料放進post
        if "message" in post:  #如果"message"在post裡面
            #print(post["message"]) #放出message的內容
            space += jieba.lcut(post["message"]) #
    res=requests.get(jd["paging"]['next']) #res=下一頁的網址
    jd=json.loads(res.text)
print(space)
```

```python
A={} #用來放文字出現幾次
for B in space: #把space的內容全部跑一遍並一個一個抓來用併定義成B
    if B in A:  #如果A裡面有B
        A[B] += 1 # 則次數+1
    else:          #否則
        A[B] = 1 #新增字詞並等於1
print(A)
```

```python
#導入
from bs4 import BeautifulSoup
#檔案路徑
path= "web-2.html"
#開啟檔案(路徑,狀態,編碼)
with open (path,"r",encoding="UTF-8") as f:
#讀取檔案並轉換成bs格式
    soup = BeautifulSoup(f.read(),"html")
#檢索第一個名為a的tag
print(soup.a.text)
#檢索第一個名為h1,屬性class為abc的tag
print(soup.find("h1",class_="abc"))
#檢索所有名為h1,屬性class為abc的tag > 輸出list
print(soup.find_all("h1",class_="abc"))
#自動縮牌以利閱讀html碼
print(soup.prettify())
print(soup.find_all("h1",[1]))
```

```python
print(soup.find_all("h1")[1]("h2")[0].text) #找所有h1的資料並選取h1裡第二個資料("h1")[1] 
                                            #再找h1裡第二個資料裡面所有有h2的資料("h1")[1]("h2")
                                            #再選取h2裡第1個資料 ("h1")[1]("h2")[0]
                                            #之後用.text輸出內容文字 ("h1")[1]("h2")[0].text
print(soup.select("h1[class=abc]"))         #選取tag 參數是class=abc 因是固定選取所有的 所以輸出會是[內容]
print(soup.select("h1.abc"))                #
```
